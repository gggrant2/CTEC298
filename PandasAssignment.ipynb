```python
#When some columns allow other values use a dictionary to be more specific
df = pd.read_csv("C:/Users/13013/Downloads/CTEC298/stock_data.csv",
                 na_values={
                'eps':['not available', 'n.a.'],
                'revenue':['not available', 'n.a.', -1],
                'people':['not available', 'n.a.']}
)
df
```

```python
#copy a file to a new file, use columns to pull specific columns
df.to_csv('new.csv', columns=['tickers','eps'])
```

```python
#read excel file
df = pd.read_excel('C:/Users/13013/Downloads/CTEC298/stock_data.xlsx', 'Sheet1')
df
```

```python
#defining function, if cell returns specific value will return new value in cell

def convert_people_cell(cell):
    if cell == 'n.a.':
        return 'sam walton'
    return cell
    
def convert_eps_cell(cell):
    if cell == 'not available':
        return 'NaN'
    return cell
    
df = pd.read_excel('C:/Users/13013/Downloads/CTEC298/stock_data.xlsx', 'Sheet1', converters = {
        'people': convert_people_cell,
        'eps': convert_eps_cell
    })
df
```

```python
#To write dataframe to excel file and customize starting row and columns
df.to_excel('new.xlsx',sheet_name='stocks', startrow=1, startcol=2)
```
